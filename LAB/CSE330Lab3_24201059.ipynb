## Creating Lagrange_Polynomial Class
General form of an $n$ degree Lagrange polynomial:

\begin{equation}
p_n(x) = \sum_{k=0}^{n} f(x_k)l(k,x) = \sum_{k=0}^{n} y_kl(k,x),\tag{1}
\end{equation}

where
\begin{equation}
l(k,x) = \prod_{j=0, j\neq k}^{n} \frac{x-x_j}{x_k-x_j}. \tag{2}
\end{equation}

Note that the Lagrange method is more efficient than the matrix method because *we do not need to calculate any inverse matrices*.
